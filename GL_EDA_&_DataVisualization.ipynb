```python
#Scatter Plot
iris= pd.read_csv('/content/drive/MyDrive/Data Science Projects/iris.csv')
iris.head()
```

```python
# working on the column titles
iris2 = iris.rename(columns=str.capitalize)
iris2.head()
```

```python
iris3 = iris2.rename(columns={'Variety': 'Species'})
iris3.head()
```

```python
# Plotting Sepal length and Petal length
sns.scatterplot(x='Sepal.length', y='Petal.length', data=iris3)
plt.show()
# There seem to be a general is a positive correlation
# The longer the sepal, the longer the petal for most species
```

```python
# Adding hue on species to discern better
sns.scatterplot(x='Sepal.length', y='Petal.length', hue='Species', data=iris3)
plt.show()
# There isn't much positve correlation in the Setosa species
# And the features are not so big
# As compared with Virginica with huge petals and sepals
```

```python
# Adding hue to Petal length to help differentiate
sns.scatterplot(x='Sepal.length', y='Petal.length', hue='Petal.length', data=iris3)
plt.show()
# the darker the data point, the bigger the value
```

```python
# Box plot
churn = pd.read_csv('/content/drive/MyDrive/Data Science Projects/Churn.csv')
churn.head()
```

```python
churn.info()
```

```python
# Plotting based on Churn and tenure columns
sns.boxplot(x='Churn', y='tenure', data=churn, hue='Churn')
# indicates a range of more customers leave when tenure is low
# there could be other factors as price and quality of service
```

```python
# Plotting based on internet service and monthly charges
sns.boxplot(x='InternetService', y='MonthlyCharges', data=churn)
# plt.show() is only used in matplotlib i think

# customers with Fiber optics IS have the highest of monthly charges
# with a customer range size almost as DSL internet users
```

```python
# Plotting based on contract
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1')
# relatively more costumers like the 1 year contract
# I think the plot may be tautological
```

```python
# Changing line width of boxplot
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1', linewidth=5)
```

```python
# Ordered costumized plotting on x axis
sns.boxplot(x='Contract', y='tenure', data=churn, palette='Set1', linewidth=5, order=['One year', 'Two year', 'Month-to-month'], hue='PaymentMethod')
```
